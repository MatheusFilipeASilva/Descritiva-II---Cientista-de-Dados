1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


previsao = pd.read_csv('Profissao- Cientista de Dados_M10_support material.csv')
print(previsao)

fig, (ax1,ax2) = plt.subplots(1, 2, figsize = (10,8))

sns.barplot(x='posse_de_veiculo', y='renda', data=previsao, estimator=np.mean, ax=ax1)
ax1.set_title('Renda por categoria de posse de veículo')

sns.barplot(x='posse_de_imovel', y='renda', data=previsao, estimator=np.mean, ax=ax2)
ax2.set_title('Renda por categoria de posse de imovel')

plt.tight_layout()
plt.show()
````

Nossos graficos nos mostram que, mesmo levando em consideração a margem de erro amostral , estas categorias são eficientes como indicadores de renda, especialmente a posse de veículo. A posse de imovel também, embora a margem de erro quase se encontre com relação ao falt
